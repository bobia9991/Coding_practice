```python
###### another example for using a custom function
# find the Pools which share common indexes.

### inspecting dataset
print("The dataset looks like: \n",pool_CR.head())

# find all duplicate values then sort rows by 'Primers' and 'Pool' column respectively.
df = pool_CR[pool_CR['Primers'].duplicated(keep=False)].sort_values(by=['Primers', 'Pool'])
print('\n ######### Step 1 ######### \n',df)

### group df by 'Primers' column then count the number of pools in each group.
print('\n ######### Step 2 ######### \n',df.groupby(by='Primers')['Pool'].value_counts())

    # as you can see, each primer group had at least two rows, as they share a common index.
    # and those rows are indexes of each group (MultiIndex)
    # sanity check
        # df.groupby(by='Primers')['Pool'].value_counts().index

### apply a custom function joining all rows of each group to one line.
    # Method 1: https://stackoverflow.com/questions/61819863/ \
            # add-index-of-value-counts-to-new-column-for-each-group-in-groupby
# f = lambda x: ' collided with '.join(x.value_counts().index)
# df2 = df.groupby(by='Primers')['Pool'].agg(f)

    # Method 2: https://stackoverflow.com/questions/36274014/ \
            # combine-rows-of-multiindex-dataframe-into-comma-separated-lists
df2 = df.groupby(by='Primers')['Pool'].agg(' collide with '.join)
print('\n ######### Step 3 ######### \n', df2)

### convert the output to a dataframe
df3 = df2.reset_index(name='Pairs of Pool')
print('\n ######### Step 4 ######### \n', df3, '\n')

### sort and unique duplicate values in the 'Pairs of pool' column
df4 = df3['Pairs of Pool'].sort_values().unique()

### finally, print the output
print('########## This is a list showing the pools having collided indexes ##########')
for i in df4:
    print(i)
```

## 4. Calculating From Data Exercise

For each of the below match-ups, choose the group that is more likely to vote for Bill Clinton. You can calculate this using the percentage of each group that intends to vote for Clinton (`vote`). Which match-up was the closest? Which had the biggest difference?

Another way to think about this: Given that a respondent is a Democrat, there is a ____ percent chance they will vote for Clinton. How does this value change if the respondent is a Republican?

1. Democrats or Republicans
2. People younger than 44 or People 44 and older
3. People who watch TV news at least 6 days a week or People who watch TV news less than 3 days a week
4. People who live somewhere with a population greater than the average respondent or People who live in a place with a population equal to or less than the average respondent

```python
# before we start, we must create columns storing the classification of sub-groups.
# for example, the classification of Democrat and Republican groups.


###### scenario 1: 
# Democrats and Republicans
# Done in the question 2.

################################################################################################

###### scenario 2: younger and older than 44
election['age_class'] = np.where(election['age'] < 44, 'above 44', 'below 44')

################################################################################################

###### scenario 3: less than 3 days or at least 6 days
def TV_times(x):
    '''
        classify respondent by watching TV times per week
    '''
    if x['TVnews'] < 3:
        return 'below 3'
    elif x['TVnews'] >= 6:
        return 'above 6'
    else:
        return '3-5'
        
election['TV_times'] = election.apply(TV_times, axis=1)

################################################################################################

###### scenario 4: less and more than the average of respondents
average_pop = election['popul'].mean()
election['pop_class'] = np.where((election['popul'] > average_pop) , 'above average', 'below average')
```

```python
# now, we create a module that calculate the percentage of each sub-group above.

def cal_percentage_and_plotting(column, group_a, group_b):
    '''
        calculates the percentage of Clinton voters in each sub-group.
    '''
    df_groupA = election[election[column] == group_a]
    Clinton_A = df_groupA[df_groupA['vote'] == 0]
    percent_Clinton_A = round((len(Clinton_A)/len(df_groupA)) * 100,2)

    df_groupB = election[election[column] == group_b]
    Clinton_B = df_groupB[df_groupB['vote'] == 0]
    percent_Clinton_B = round((len(Clinton_B)/len(df_groupB)) * 100,2)

    '''
        combines those values into a dataframe
    '''
    data = pd.DataFrame()
    data['Group'] = [group_a, group_b]
    data['percentage'] = [percent_Clinton_A, percent_Clinton_B]

    '''
        plots a bar chart illustrating those percentages.
    '''
    p = (ggplot(data)
                    + aes(x= 'Group', y = 'percentage', fill = 'Group') 
                    + geom_bar(stat = 'identity', position='dodge2')
                    + geom_text(aes(label='percentage'),
                            position = position_dodge(width = 0.9),
                            va = 'bottom')
    )
    return p
```

```python
###### finally, plot all of figures in the same area by the patchworklib package.

# assign variable for each figure.
p1 = cal_percentage_and_plotting('party', 'Democrat', 'Republican')
p2 = cal_percentage_and_plotting('age_class', 'below 44', 'above 44')
p3 = cal_percentage_and_plotting('TV_times', 'below 3', 'above 6')
p4 = cal_percentage_and_plotting('pop_class', 'below average', 'above average')

# plot
p1 = pw.load_ggplot(p1, figsize=(2,3))
p2 = pw.load_ggplot(p2, figsize=(2,3))
p3 = pw.load_ggplot(p3, figsize=(2,3))
p4 = pw.load_ggplot(p4, figsize=(2,3))
p1234 = (p1|p2)/(p3|p4)
p1234.savefig()
```
